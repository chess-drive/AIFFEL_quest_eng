```python
'''
Created on 2026. 1. 16.

@author: chaliepark
'''

import random
import time

## Character클래스
class Character():
    def __init__(self,name,level,hp,attack,armour):
        self.chrctr_info={"name":name,"level":level,"hp":hp,"attack":attack,"armour":armour,"is_alive":True}

    def is_alive(self):
        return self.chrctr_info["is_alive"]
    def take_damage(self,damage):
        if self.chrctr_info["armour"] <=damage:
            self.chrctr_info["hp"]-= damage
            if self.chrctr_info["hp"] <=0:
                self.chrctr_info["is_alive"]=False

    def attack_target(self):
        return random.randint(1,self.chrctr_info["attack"]) ##데미지는 1~공격력사이의 값.


## Player클래스
class Player(Character):
    def __init__(self,name):
        super().__init__(name,1,100,25,5)
        self.exp=0
    def gain_experience(self,exp_point):
        self.exp += exp_point
    def level_up(self):
        self.chrctr_info["level"] +=1
        self.chrctr_info["attack"] +=10
        self.chrctr_info["armour"] +=5
        self.exp=0

class Monster(Character):
    def __init__(self,name,level):
        #- 체력: 10~30 사이의 랜덤한 정수 * 레벨
        #- 공격력: 5~20 사이의 랜덤한 정수 * 레벨
        #- 방어력: 1~5 사이의 랜덤한 정수 * 레벨
        super().__init__(name,level,random.randint(10,30)*level,random.randint(5,20)*level,random.randint(1,5)*level)


def battle(player,monster):
    battle_bool=True
    return_message=""
    while battle_bool:
        # player턴
        player_damage=player.attack_target()
        monster.take_damage(player_damage)
        print(player.chrctr_info["name"],"이 ",monster.chrctr_info["name"],"에게 ",str(player_damage)," 만큼 공격했다!")
        print(monster.chrctr_info["name"],"의 체력:",str(monster.chrctr_info["hp"]))

        if player.is_alive() and not monster.is_alive():
            return_message=monster.chrctr_info["name"]+"을 이겼다!\n전투 승리!"
            battle_bool=False

        # monster턴
        monster_damage=monster.attack_target()
        player.take_damage(monster_damage)
        print(monster.chrctr_info["name"],"이 ",player.chrctr_info["name"],"에게 ",str(monster_damage)," 만큼 공격했다!")
        print(player.chrctr_info["name"],"의 체력:",str(player.chrctr_info["hp"]))

        if monster.is_alive() and not player.is_alive():
            return_message=player.chrctr_info["name"]+"가 죽었다.\n전투 패배.."
            battle_bool=False
    print(return_message)

def main():
    monster_dict={"슬라임":1,"고블린":2,"오크":3}
    park = Player(input("player 이름을 입력하세요:"))
    for key,val in monster_dict.items():
        battle(park,Monster(key,val))
        if not park.chrctr_info["is_alive"]:
            break
        print('\n+'*3)
        time.sleep(1)



main()



```
